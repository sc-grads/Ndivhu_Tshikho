```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset('tips')

# Create histogram
sns.displot(tips['total_bill'], kde=False, bins=10)

# Show the plot
plt.show()  # Add parentheses to call the function
```

```python
sns.jointplot(x='total_bill',y='tip', data=tips, kind = 'kde')
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.rugplot(tips['total_bill'])
plt.show()
```

```python
sns.kdeplot(tips['total_bill'])
plt.show()
```

```python
#Categorical Plots
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
plt.show()
```

```python
sns.countplot(x='sex', data=tips)
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue= 'smoker')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue = 'sex', split = True)
plt.show()
```

```python
sns.stripplot(x='day', y = 'total_bill', data = tips, jitter=True, hue='sex')
plt.show()
```

```python
sns.swarmplot(x='day', y='total_bill', data = tips, hue = 'sex')
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin')
plt.show()
```

```python
# Load example dataset
flights = sns.load_dataset('flights')

# Create pivot table with observed=False to suppress the warning
pivot_table = flights.pivot_table(index='month', columns='year', values='passengers', observed=False)
```

```python
pivot_table
```

```python
sns.heatmap(pivot_table, cmap='magma', linecolor='white', linewidths=1)
plt.show()
```
