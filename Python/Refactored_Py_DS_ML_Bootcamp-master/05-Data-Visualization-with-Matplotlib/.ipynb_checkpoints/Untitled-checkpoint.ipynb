```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,.5,11)
y = x *82
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.show()
plt.subplot(1,2,2)
plt.plot(y,x, 'b')
plt.show()
```

```python
#OO
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for currentax in axes:
    currentax.plot(x,y)

fig
```
